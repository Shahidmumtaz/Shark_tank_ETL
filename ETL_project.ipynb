```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# load csv file and store into a DataFrame
load_csv = "Resources/shark_tank_companies.csv"
shark_tank_1 = pd.read_csv(load_csv)
shark_tank_1.head()
```

```python
# Delete duplicates
shark_tank_1.drop_duplicates(keep="first", inplace=True)
```

```python
# sort the data in the dataframe by valuation
shark_tank_1.sort_values(["Valuation"], ascending = False, inplace = True)
```

```python
# Create a filtered dataframe from specific columns
shark_tank_1_cols = ["Season", "Deal", "Category", "Valuation", "Shark1", "Shark2", "Shark3", "Shark4", "Shark5"]
shark_tank1_transformed= shark_tank_1[shark_tank_1_cols].copy()

# Remove null values
shark_tank1_transformed = shark_tank1_transformed[shark_tank1_transformed['Deal'].notnull()]  

shark_tank1_transformed.head()
```

```python
# store csv file into a Dataframe
load_csv_file = "Resources/sharktank.csv"
shark_tank_2 = pd.read_csv(load_csv_file)
shark_tank_2.head()
```

```python
# Create a filtered dataframe from specific columns
shark_tank_2_cols = ["Season", "Deal", "Category", "Valuation", "Sharks"]
shark_tank2_transformed= shark_tank_2[shark_tank_2_cols].copy()

# Define a cleaned Dataframe with only relevant values columns and remove duplicates
shark_tank2_transformed.sort_values("Valuation", ascending=True, inplace=True)
shark_tank2_transformed.drop_duplicates(keep="first", inplace=True)

shark_tank2_transformed[shark_tank2_transformed['Deal'].notnull()]  

shark_tank2_transformed.head()
```

```python
# merge the two dataframes
shark_tank_merged = pd.merge(shark_tank1_transformed, shark_tank2_transformed, on=["Season", "Deal", "Category"], how="left")
# keep the necessary columns
shark_tank_merged_cols = ["Season", "Deal", "Category", "Valuation_x", "Shark1", "Shark2", "Shark3", "Shark4", "Shark5"]
shark_tank_merged_final= shark_tank_merged[shark_tank_merged_cols].copy()
shark_tank_merged_final.head()
```

```python
# change column names
shark_tank_final = shark_tank_merged_final.rename(index=str,columns={"Valuation_x":"Valuation"},inplace=True)
shark_tank_final.head()
```
